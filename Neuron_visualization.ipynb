```python
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
```

```python
X = np.zeros((30, 2))
for i in range(30):
    X[i] = [random.randint(1,100),random.randint(1,100)]
#X = np.array([[0, 1], [1, 0], [3, 2], [5,3], [-1,2]])
print(X)
```

```python
plt.plot(X[:,0],X[:,1], 'ro')
plt.show()
```

```python
distances = np.zeros((len(X), len(X)))
print (distances)
```

```python
#находим N (случайно) ближайших связей
connections = np.zeros((len(X), len(X)))
for i in range(len(X)):
    connections_count = random.randint(1,len(X)-10)
    mins = np.zeros(connections_count)
    argmin_s = np.zeros(connections_count)
    for j in range(i,len(X)):
        distances[i][j] = np.linalg.norm(X[i]-X[j]) 
        if min(mins)==0:
            argmin_s[np.argmin(mins)] = j
            mins[np.argmin(mins)] = distances[i][j]
        elif (distances[i][j]<max(mins)):
            argmin_s[np.argmax(mins)] = j
            mins[np.argmax(mins)] = distances[i][j]
    print(argmin_s)
    for j in argmin_s:
        if j!=0:
            connections[i][int(j)]=1   
        
   # print(connections_count)
print(distances)
print(connections)
```

```python
#рисовалка связей
for i in range(len(connections)):
    for j in range(len(connections)):
        if connections[i][j]==1:
            plt.plot([X[i][0],X[j][0]],[X[i][1],X[j][1]], 'bo-')
plt.plot(X[:,0],X[:,1], 'ro')
plt.show()
```

```python
cov = X.T.dot(X)
print(cov)
```

```python
probabilities = np.zeros((len(X), len(X)))
for i in range(len(X)):
    for j in range(len(X)):
        if connections[i][j]==1 or i==j:
            probabilities[i][j] = multivariate_normal.pdf(X[j], X[i], cov)
        else:
            probabilities[i][j] = 0        
    probabilities[i] = probabilities[i]/ probabilities[i][i]
    probabilities[i][i] = 0
    
print (probabilities)
```

```python
actives_neuron = np.zeros(len(X))
actives_neuron[0] = 1
for k in range(30):
    if k%10==0:
        actives_neuron[0] = 1
    probabilities_n =np.multiply(probabilities.T,actives_neuron).T
    actives_neuron = np.zeros(len(X))
    #print (probabilities_n)
    for i in range(len(connections)):
        for j in range(len(connections)):
            if connections[i][j]==1:
                plt.plot([X[i][0],X[j][0]],[X[i][1],X[j][1]], 'bo-')
                if  random.random()<probabilities_n[i][j]:
                    actives_neuron[j] = 1
    for j in range(len(actives_neuron)):
        if actives_neuron[j]>0:
            plt.plot(X[j,0],X[j,1],'ro')
    #print (actives_neuron)
    plt.show()
```
